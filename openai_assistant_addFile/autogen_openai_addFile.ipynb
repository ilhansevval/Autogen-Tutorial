## **Import Libraries**

```python
import autogen
from autogen.agentchat.contrib.gpt_assistant_agent import GPTAssistantAgent
import openai
```

## **Initialize Client**

```python
import os
os.environ["OPENAI_API_KEY"] = "secret_key"
```

```python
client = openai.OpenAI()
```

## **Add File to OpenAI Assistant**

```python
file = client.files.create(
    file=open("/content/data.csv", "rb"),
    purpose='assistants'
)
```

## **Create OpenAI Assistant**

```python
assistant = client.beta.assistants.create(
    name="Python Developer",
    instructions="You are a python developer",
    file_ids=[file.id],
    tools=[
        {
            "type": "code_interpreter",
            "config": {
            "language": "python",
            "environment": "python3.8",
            "timeout": 10,
            "max_memory": "512MB"
        }

        }
    ],
    model="gpt-3.5-turbo"
)
```

## **Create a Config List**

```python
config_list = [
    {
        "model" : "gpt-4-1106-preview",
        "api_key" : "secret_key"
    },
    {
        "model" : "gpt-4-vision-preview",
        "api_key" : "secret_key"
    }
]
```

```python
llm_config = {
    "assistant_id": assistant.id,
    "tools": [ { "type": "code_interpreter" } ],
    "file_ids": [file.id]
}
```

```python
gpt_assistant = GPTAssistantAgent(
    name="assistant",
    llm_config=llm_config,
    instructions="""
    You are an expert at writing python code to solve problems.
    """,)
```

```python
user_proxy = autogen.UserProxyAgent(
    name="user_proxy",
    human_input_mode= "NEVER",
    is_termination_msg=lambda msg: "TERMINATE" in msg["content"],
    code_execution_config={
        "work_dir": "coding",
        "use_docker": False,
    },
   max_consecutive_auto_reply=10,
   system_message = "Reply TERMINATE if the task has been solved for your full satisfaction. Or reply CONTINUE if the task is not solved yet")
```

```python
user_proxy.initiate_chat(
    gpt_assistant,
    message="""
    Give me the overview of data
    """
)
```
